# Análise Inicial

A análise das campanhas de marketing digital para a marca Wizard by Pearson revela insights importantes sobre o desempenho das estratégias implementadas no Google Search e no YouTube. Os dados quantitativos obtidos demonstram uma discrepância significativa entre as duas plataformas em termos de eficácia e resultados.

A campanha realizada no Google Search, identificada como "[Wizard ON] Google | Search | Help me Buy | Custom", apresentou um desempenho notável. Com um total de 136,185 interações, que se traduziram em 136,185 cliques, a campanha alcançou uma taxa de conversão de 12.932%. Este valor não só reflete a eficácia da campanha em motivar os usuários a realizar ações desejadas, mas também destaca a qualidade do engajamento gerado. Além disso, o custo médio por clique foi de aproximadamente $7.49, um investimento que se justifica pelo alto retorno em conversões, evidenciando uma gestão de custos eficiente diante dos resultados obtidos.

Por outro lado, a campanha veiculada no YouTube, rotulada como "[wizard on] youtube | tráfego | help me choose...", não replicou o sucesso da campanha no Google Search. Ela falhou em registrar qualquer interação, clique ou conversão, resultando em uma taxa de conversão de 0%. Este resultado sugere que a campanha pode não ter sido efetivamente configurada ou que a abordagem e o conteúdo não ressoaram com o público-alvo no YouTube. Isso evidencia uma necessidade crítica de reavaliação da estratégia no YouTube, considerando uma revisão no conteúdo, segmentação de público ou mesmo na metodologia de execução da campanha.

Estas análises apontam para a importância de uma avaliação contínua e ajustes nas estratégias de marketing digital. Enquanto a campanha no Google Search demonstra uma alocação de orçamento eficaz e um retorno sobre o investimento satisfatório, a campanha no YouTube destaca a necessidade de uma intervenção estratégica para corrigir falhas ou melhorar a execução. Ajustar táticas conforme necessário pode não apenas melhorar a eficiência das campanhas em plataformas com desempenho inferior mas também maximizar o impacto geral das iniciativas de marketing digital da marca Wizard by Pearson.

```python
descriptive_stats = ga_data[['Users', 'Sessions']].describe()
print(descriptive_stats)
```

As estatísticas descritivas das colunas "Users" e "Sessions" do conjunto de dados revelam uma discrepância significativa no tráfego de diferentes páginas do site. Enquanto a média de usuários por página é de aproximadamente 1206, o desvio padrão alto (7131) indica uma variabilidade significativa, com algumas páginas atraindo até 85604 usuários, contrastando com um mínimo de apenas 11 usuários. Similarmente, as sessões variam de 0 a 94025, com uma média de 1115 sessões por página. A maioria das páginas tem tráfego extremamente baixo, como evidenciado pela mediana de 11 para ambas as métricas, indicando que poucas páginas de alto desempenho estão elevando a média. Este perfil sugere uma análise mais aprofundada das estratégias de conteúdo e marketing das páginas de alto desempenho, ao mesmo tempo que se considera a otimização das páginas de baixo tráfego para melhorar seu desempenho global.








```python
top_pages = ga_data.nlargest(10, 'Users')
sns.set(style="whitegrid")

plt.figure(figsize=(14, 8))
sns.barplot(data=top_pages, x='Page Title', y='Users', color='blue', label='Usuários')
sns.barplot(data=top_pages, x='Page Title', y='Sessions', color='red', alpha=0.6, label='Sessões')

plt.xlabel('Título da Página', fontsize=12)
plt.ylabel('Contagem', fontsize=12)
plt.title('Top 10 Páginas por Número de Usuários e Sessões - 2023', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend()
plt.tight_layout()
plt.show()
```

O gráfico de barras que analisa as 10 páginas mais populares por número de usuários e sessões em 2023 revela que a página "Wizard On" se destaca significativamente, indicando sua eficácia em atrair tráfego. As páginas com conteúdo promocional e cursos específicos dominam o topo da lista, sugerindo um alto engajamento nesses temas. Nota-se uma discrepância entre o número de usuários e sessões em algumas páginas, o que pode indicar visitas recorrentes ou um engajamento prolongado. Essa análise sugere que páginas de alto desempenho podem beneficiar-se ainda mais de estratégias de marketing direcionadas, enquanto páginas com desempenho inferior podem precisar de ajustes em conteúdo ou abordagem para aumentar sua eficácia e relevância.
